```python
import pandas as pd
import os
import shutil
```

```python
File_path = "C:/Users/sanju/Desktop/dee/covid-chestxray-dataset-master/covid/metadata.csv"
Image_path = "C:/Users/sanju/Desktop/dee/covid-chestxray-dataset-master/covid/images"
```

```python
df = pd.read_csv(File_path)
print(df.shape)
```

```python
df.head()
```

```python
Target_dir = "Dataset/covid"

if not os.path.exists(Target_dir):
    os.mkdir(Target_dir)
    print ("covid folder created")
```

```python
cnt = 0

for (i, row) in df.iterrows():
    if row["finding"] == "Pneumonia/Viral/COVID-19" and row["view"]== "PA" or row["finding"]== "Pneumonia/Viral/SARS":
        filename = row["filename"]
        image_path = os.path.join(Image_path, filename)
        image_copy_path = os.path.join(Target_dir, filename)
        shutil.copy2(image_path, image_copy_path)
        print("data copy",i)
        cnt += 1
        
print (cnt)
```
